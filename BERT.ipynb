```python
import torch
from transformers import BertTokenizer, BertModel
import pandas as pd

# Load your DataFrame from the CSV file


# Load pre-trained BERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

# Function to generate BERT embeddings for text
def generate_bert_embeddings(text):
    # Tokenize the text
    input_ids = tokenizer.encode(text, add_special_tokens=True, max_length=512, truncation=True, padding='max_length')
    
    # Convert tokens to PyTorch tensors
    input_ids = torch.tensor(input_ids).unsqueeze(0)  # Batch size 1
    
    # Create attention mask
    attention_mask = torch.ones(input_ids.shape, dtype=torch.long)
    
    # Forward pass through BERT model
    with torch.no_grad():
        outputs = model(input_ids, attention_mask=attention_mask)
    
    # Extract the embeddings from the model outputs
    embeddings = outputs.last_hidden_state.mean(dim=1).squeeze().numpy()
    
    return embeddings

# Apply the function to generate embeddings for each review text


# Now, df['text_embeddings'] contains the BERT embeddings for each review text
```

```python
df = pd.read_csv('allReviews.csv')
import numpy as np
# Get the columns of the DataFrame
columns = df.columns

# Convert DataFrame to numpy array
data_np = df.to_numpy()

# Get the index of the column containing text data
text_column_index = df.columns.get_loc('reviewText')  # Assuming 'reviewText' is the column containing text data

# Generate BERT embeddings for each text in the DataFrame

i = 0

# List to store embeddings
embeddings_list = []

# Iterate over each row in the numpy array
for row in data_np:
    # Extract text from the row
    text = row[text_column_index]
    
    # Generate BERT embeddings for the text
    embeddings = generate_bert_embeddings(text)
    
    # Concatenate embeddings with the rest of the row data
    combined_data = np.concatenate((row, embeddings), axis=None)
    
    # Append combined data to embeddings list
    embeddings_list.append(combined_data)
    
    # Print progress (optional)
    if i % 100 == 0:
        print(i)
    
    # Increment counter
    i += 1

# Convert embeddings list to numpy array
data_with_embeddings_np = np.array(embeddings_list)

# Save the NumPy array to a .npy file
np.save('data_with_embeddings.npy', data_with_embeddings_np)
```
